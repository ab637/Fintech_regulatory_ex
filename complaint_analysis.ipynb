```python
import pandas as pd
import matplotlib.pyplot as plt

# Step 1: Load your full complaint dataset from CSV
df = pd.read_csv("monthly_complaints_data.csv")  # Replace with your actual file path

# Step 2: Convert 'Month' column to datetime
df['Month'] = pd.to_datetime(df['Month'], format='%b-%y')

# Step 3: Plot total complaints over time
plt.figure(figsize=(8, 4))
plt.plot(df['Month'], df['Total_Complaints'], marker='o', color='darkblue')
plt.title('Total Complaints Over Time')
plt.xlabel('Month')
plt.ylabel('Total Complaints')
plt.grid(True)
plt.tight_layout()
plt.show()

# Step 4: Plot individual complaint categories over time
plt.figure(figsize=(10, 5))
plt.plot(df['Month'], df['Interest_Rate_Complaints'], marker='o', label='Interest Rate')
plt.plot(df['Month'], df['Recovery_Agent_Threats'], marker='o', label='Recovery Threats')
plt.plot(df['Month'], df['No_Response_Customer_Care'], marker='o', label='No Response')
plt.plot(df['Month'], df['Other_Categories'], marker='o', label='Other Categories')
plt.title('Category-wise Complaint Trends')
plt.xlabel('Month')
plt.ylabel('Number of Complaints')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

# Step 5: Stacked bar chart of complaint category composition
df.set_index('Month', inplace=True)
df[['Interest_Rate_Complaints', 'Recovery_Agent_Threats', 
    'No_Response_Customer_Care', 'Other_Categories']].plot(
    kind='bar', stacked=True, figsize=(10, 6), colormap='Paired'
)
plt.title('Monthly Complaint Composition by Category')
plt.xlabel('Month')
plt.ylabel('Complaint Count')
plt.tight_layout()
plt.show()
```
